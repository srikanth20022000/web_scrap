```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import re

college_names=[]

course=[]

place=[]
owner=[]
rating=[]
fees=[]
exams=[]
other_courses=[]






for i in range(1,11):
    response=requests.get(f"https://engineering.careers360.com/colleges/list-of-engineering-colleges-in-india?page={i}&sort_by=4")
    response

    response.text
    data=BeautifulSoup(response.text)
    data

# college names
    
    for i in range(len(data.find_all("h3",class_="college_name"))):
        extrcted_text=data.find_all("h3",class_="college_name")[i].text
        place_extracted=re.findall(pattern=r".*, ",string=extrcted_text)
        college_names.append(place_extracted)
    college_names

# course 
    for i in range(len(data.find_all("a",class_="general_text"))):
    
        course.append(data.find_all("a",class_="general_text")[i].text)
    course

#place
    for i in range(len(data.find_all("div",class_="content_block d-none d-md-block d-md-flex flex-row justify-content-between"))):
        extrcted_text=data.find_all("div",class_="content_block d-none d-md-block d-md-flex flex-row justify-content-between")[i].text
        place_extracted=re.findall(pattern=r".*\w+,",string=extrcted_text)
        place.append(place_extracted)
    place

#owner
    for i in range(len(data.find_all("div",class_="content_block d-none d-md-block d-md-flex flex-row justify-content-between"))):
        extrcted_text=data.find_all("div",class_="content_block d-none d-md-block d-md-flex flex-row justify-content-between")[i].text
        private_extracted=re.findall(pattern=r"Private|Public",string=extrcted_text)
        owner.append(private_extracted)
    owner

# rating
    for i in range(len(data.find_all("div",class_="content_block d-none d-md-block d-md-flex flex-row justify-content-between"))):
        extrcted_text=data.find_all("div",class_="content_block d-none d-md-block d-md-flex flex-row justify-content-between")[i].text
        star_extracted=re.findall(pattern=r"  \d+..",string=extrcted_text)
        if len(star_extracted) > 0:
            rating.append(star_extracted)
        else:
            rating.append(np.nan)
    rating

# fees

    for i in range(len(data.find_all("ul",class_="snippet_list"))):
        extrcted_text=data.find_all("ul",class_="snippet_list")[i].text
        star_extracted=re.findall(pattern=r"â‚¹.* Lakhs",string=extrcted_text)
        if len(star_extracted) > 0:
            fees.append(star_extracted)
        else:
            fees.append(np.nan)
    fees

# exams

    for i in range(len(data.find_all("ul",class_="snippet_list"))):
        extrcted_text=data.find_all("ul",class_="snippet_list")[i].text
        star_extracted=re.findall(pattern=r"Exams :.\w+",string=extrcted_text)
        if len(star_extracted) > 0:
            exams.append(star_extracted)
        else:
            exams.append(np.nan)
    exams

# other courses

    for i in range(len(data.find_all("ul",class_="snippet_list"))):
        extrcted_text=data.find_all("ul",class_="snippet_list")[i].text
        star_extracted=re.findall(pattern=r"B.Tech..\d+ Courses",string=extrcted_text)
        if len(star_extracted) > 0:
            other_courses.append(star_extracted)
        else:
            other_courses.append(np.nan)
    other_courses
```

```python
college_names=college_names[0:1001:2]
college_names
```

```python
eng_coll={"names":college_names,
          "Top_course":course,
          "location":place,
          "public/private":owner,
          "rating/5":rating,
          "fee_structure":fees,
          "entrance_text":exams,
          "other_course":other_courses
         }
          
```
