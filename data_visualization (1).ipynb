```python
import pandas as pd
```

```python
mat=pd.read_csv(r"C:\Users\srika\Downloads\college_data.csv")
mat
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

# univarient analysis

# numerical

```python
# histogram , box plot
# fee structure
plt.figure(figsize=(3,2), dpi=250)
plt.hist(mat['fee_structure'], bins=20, edgecolor="black",)
plt.grid(linestyle="--", color="red", linewidth=0.4)
plt.xlabel('fee')
plt.ylabel('Count/colleges')
plt.title("based on fees no of colleges")
plt.show()
```

```python
#rating
plt.figure(figsize=(3,2), dpi=250)
plt.hist(mat['rating/5'], bins=20, edgecolor="black",)
plt.grid(linestyle="--", color="red", linewidth=0.4)
plt.xlabel('rating')
plt.ylabel('colleges/count')
plt.title("based on rating no of colleges")
plt.show()
```

```python
sns.boxplot(data=mat, y='fee_structure', width=0.7)
plt.axhline(5.8, linestyle="--", color="red")
plt.title("outliers in fee structure")
plt.show()
```

```python
plt.title("outliers in other courses")
sns.boxplot(data=mat, y='other_course', width=0.7)
plt.axhline(5.8, linestyle="--", color="red")
plt.show()
```

```python
#courses
plt.figure(figsize=(3,2), dpi=250)
plt.hist(mat['other_course'], bins=20, edgecolor="black",)
plt.grid(linestyle="--", color="red", linewidth=0.4)
plt.xlabel('courses')
plt.ylabel('colleges')
plt.title("colleges having other courses")
plt.show()
```

# category

```python
sns.countplot(data = mat, y="entrance_text")
plt.title("entrance test to get admission in engineering")
plt.show()
```

```python
# category
from wordcloud import WordCloud
```

```python
text = mat["entrance_text"].str.cat(sep=" ")
text
```

```python
text =mat["entrance_text"].str.cat(sep=" ")
wordcloud = WordCloud(max_font_size=80, max_words=50, colormap='flag', width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(10,10))
plt.plot()
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("main exam in engineering college")
plt.show()
```
